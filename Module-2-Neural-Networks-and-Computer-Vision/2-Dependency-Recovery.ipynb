**Задание 1.**

Для того чтобы показать, что решение задачи оптимизации совпадает с решением системы линейных уравнений $Xa=y$, если система разрешима, найдем производную функции потерь по параметрам $a$ и приравняем её к нулю.

   $$L(a) = \|Xa - y\|_2^2 = (Xa - y)^T (Xa - y).$$

   $$L(a) = (Xa)^T Xa - 2y^T Xa + y^T y.$$

   $$\frac{\partial L}{\partial a} = 2X^T X a - 2X^T y.$$

   $$2X^T X a - 2X^T y = 0.$$

   $$X^T X a = X^T y.$$

$$M + 1 >= N + 1$$

$$M >= N$$
Таким образом, степень полинома М должна удовлетворять неравенству, $$М >= N$$ чтобы он мог точно пройти через все точки.

**Задание 5.**



В случае, когда $ M > N $, матрица $ X $ (размерности $(N + 1) × (М + 1)$) имеет больше столбцов, чем строк. Это значит, что она не имеет полного ранга и, следовательно, существует ненулевое пространство решений, называемое ядром или нулевым пространством матрицы $ X $.

Ядро матрицы $ X $ можно определить через собственные векторы матрицы $ X^T X $, соответствующие нулевым собственным значениям. Эти векторы составляют пространство решений, добавление которых к любому решению задачи оптимизации даёт новое решение.
Итак, матрица, собственные векторы которой, соответствующие её нулевым собственным числам, образуют ядро матрицы $ Х $, - это матрица $ X^T X $.

**Задание 6.**

В случае, когда $ M > N $, матрица $ X $ (размерностью $ (N + 1) × (M + 1) $) имеет больше столбцов, чем строк. Это означает, что её ранг ограничен числом строк, то есть максимум равен $ N + 1 $.

Размерность нулевого пространства (ядра) матрицы X равна числу линейно независимых векторов, которые составляют пространство решений однородного уравнения $ Xz = 0 $. Она рассчитывается по формуле:
$$ Размерность (ядра) = (M + 1) - Ранг(Х) $$
Так как ранг матрицы X не может превышать N + 1, получаем:
$$ Размерность (ядра) = (M + 1) - (N + 1) = M - N $$
